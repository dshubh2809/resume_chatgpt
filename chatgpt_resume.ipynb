```python
import fitz  # PyMuPDF
import re
import json

def extract_resume_text(pdf_path):
    document = fitz.open(pdf_path)
    text = ""
    for page_num in range(len(document)):
        page = document.load_page(page_num)
        text += page.get_text("text")
    return text

def parse_resume(text):
    resume = {}
    
    # Extract basic details
    resume['name'] = re.search(r'(.*)\n\+\d{2}-\d{10} \|', text).group(1).strip()
    resume['phone'] = re.search(r'\+(\d{2}-\d{10}) \|', text).group(1).strip()
    resume['email'] = re.search(r'\| (.*?) \|', text).group(1).strip()
    resume['github'] = re.search(r'\| (https://github.com/.*?) \|', text).group(1).strip()
    resume['linkedin'] = re.search(r'\| (https://www.linkedin.com/in/.*?)\n', text).group(1).strip()
    
    # Extract education details
    education_text = re.search(r'EDUCATION(.*?)(?=EXPERIENCE)', text, re.DOTALL).group(1).strip()
    education = [edu.strip() for edu in education_text.split('\n') if edu.strip()]
    resume['education'] = education
    
    # Extract experience details
    experience_text = re.search(r'EXPERIENCE(.*?)(?=SKILLS)', text, re.DOTALL).group(1).strip()
    experience = []
    current_exp = {}
    for exp in experience_text.split('\n'):
        if exp.strip():
            if '•' in exp:
                if 'details' not in current_exp:
                    current_exp['details'] = []
                current_exp['details'].append(exp.strip())
            else:
                if current_exp:
                    experience.append(current_exp)
                current_exp = {'role': exp.strip()}
    if current_exp:
        experience.append(current_exp)
    resume['experience'] = experience
    
    # Extract skills
    skills_text = re.search(r'SKILLS(.*?)(?=Certifications)', text, re.DOTALL).group(1).strip()
    skills = [skill.strip() for skill in skills_text.split('\n') if skill.strip()]
    resume['skills'] = skills
    
    # Extract projects
    projects_text = re.search(r'PROJECTS(.*?)(?=Legal AI)', text, re.DOTALL).group(1).strip()
    projects = []
    current_proj = {}
    for proj in projects_text.split('\n'):
        if proj.strip():
            if '•' in proj:
                if 'details' not in current_proj:
                    current_proj['details'] = []
                current_proj['details'].append(proj.strip())
            else:
                if current_proj:
                    projects.append(current_proj)
                current_proj = {'title': proj.strip()}
    if current_proj:
        projects.append(current_proj)
    resume['projects'] = projects
    
    return resume

pdf_path = r'C:\Users\user\Downloads\New folder (3)\My_resume.pdf'  # Using raw string literal
resume_text = extract_resume_text(pdf_path)
resume_json = parse_resume(resume_text)
print(json.dumps(resume_json, indent=2))
```
